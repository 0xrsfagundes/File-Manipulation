```python
import pandas as pd
import openpyxl

def read_excel_tabs(file_path):
    # Load the Excel workbook
    workbook = openpyxl.load_workbook(file_path)
    
    # Initialize an empty dictionary to store data from each tab
    all_data = {}
    
    # Loop through all the tabs
    for sheet_name in workbook.sheetnames:
        # Read data from each tab into a DataFrame
        df = pd.read_excel(file_path, sheet_name=sheet_name)
        
        # Store the DataFrame in the dictionary with the tab name as the key
        all_data[sheet_name] = df
    
    # Combine all DataFrames into a single DataFrame
    combined_df = pd.concat(all_data, ignore_index=True)
    
    return combined_df

# Example usage:
file_path = "xls/B2B Audiences.xlsx"  # Replace this with the path to your Excel file
combined_data = read_excel_tabs(file_path)
combined_data
```

```python
combined_data.to_csv('xls/combined_data.csv')
```
