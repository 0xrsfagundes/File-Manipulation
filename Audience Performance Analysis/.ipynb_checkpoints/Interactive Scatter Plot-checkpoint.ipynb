```python
import warnings
import pandas as pd
import numpy as np
import plotly.express as px

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
csv_file = "csv/grouped_csb_df_export.csv"
df = pd.read_csv(csv_file)
```

```python
# Assuming df is your DataFrame
df['Spend range'] = np.select(
    [
        (df['Spend'] < 20000),                  # Spend < 20K
        (df['Spend'] >= 20000) & (df['Spend'] < 100000),  # Spend >= 20K and < 100K
        (df['Spend'] >= 100000)                 # Spend >= 100K
    ],
    [
        'Spend < 20K',
        'Spend >= 20K and < 100K',
        'Spend >= 100K'
    ],
    default='Other'
)
```

```python
df.head()
```

```python
# Create scatter plot using Plotly Express with hover_name
fig = px.scatter(df, x='ROAS', y='Spend', color='Funnel', title='FY24Q4 AdTech 3P to 1P Opportunity Analysis - CSB ROAS Analysis', hover_name='Campaign Name', facet_col="Spend range")
fig.update_traces(marker=dict(size=12, opacity=0.7))

# Export the plot as an HTML file
fig.write_html('html/FY24Q4 AdTech 3P to 1P Opportunity Analysis - CSB ROAS Analysis.html')
```
