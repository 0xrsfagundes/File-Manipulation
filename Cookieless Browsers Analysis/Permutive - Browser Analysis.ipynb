```python
df_filter = pd.DataFrame(df[df['Campaign'].str.contains('2024_Q2_US_CSBO-CON_AP_DISPLAY_PROSPECTING_NA')])

df_permutive = df_filter[df_filter['Placement'].str.contains('Permutive|Permutvie|permu|Permu|ermu')]

df_non_permutive = df_filter[~df_filter['Placement'].str.contains('Permutive|Permutvie|permu|Permu|ermu')]
```

```python
df_permutive_grouped = df_permutive.groupby(['Browser/Platform'])['Impressions'].sum().reset_index()

total_1 = df_permutive_grouped['Impressions'].sum()

df_permutive_grouped['Percentage'] = df_permutive_grouped['Impressions'] / total_1 * 100

df_permutive_grouped = df_permutive_grouped.sort_values(by='Impressions', ascending=False)

df_permutive_grouped
```

```python
df_non_permutive_grouped = df_non_permutive.groupby(['Browser/Platform'])['Impressions'].sum().reset_index()

total_1 = df_non_permutive_grouped['Impressions'].sum()

df_non_permutive_grouped['Percentage'] = df_non_permutive_grouped['Impressions'] / total_1 * 100

df_non_permutive_grouped = df_non_permutive_grouped.sort_values(by='Impressions', ascending=False)

df_non_permutive_grouped
```
