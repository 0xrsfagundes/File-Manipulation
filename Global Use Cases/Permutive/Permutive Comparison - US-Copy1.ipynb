```python
import warnings
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
warnings.filterwarnings('ignore')
```

```python
folder_path = r'C:\Users\Rafael_Fagundes\Downloads\delve_compiled_data.csv'

df = pd.read_csv(folder_path)
```

```python
# Filter the dataset by US, 2024Q3 and 3P
df_filtered = df[
                (df['Fiscal Quarter'].isin(['2024-Q2']))
                 & (df['Country'] == 'United Kingdom')
                 #& (df['Audience Type'] == '3PD') 
                 #& (df['BU'] == 'CSB') 
                 #& (df['Display Dell Vehicle Mapped'] == 'Display')
                 #& (df['Display Funnel Mapped'] == 'Active Prospects')
                ]
```

```python
# Create a flag for the main Vendors
df_filtered['Permutive_flag'] = df_filtered['Media Buy Name'].str.contains(r'Permutive|Permutvie|permu|Permu|ermu', case=False)
```

```python
df_clean = df_filtered[(df_filtered['Clicks'] <= df_filtered['Impressions'])
                      & (df_filtered['Spend'] > 0)
                      ]
```

```python
df_clean['Audience Type Name v2'] = df_filtered['Audience Type Name'] + np.where(df_filtered['Permutive_flag'], ' - Permutive', '')
```

```python
df_clean['CTR'] = (df_clean['Clicks'] / df_clean['Impressions']) * 100
df_clean['CPM'] = 1000 * df_clean['Spend'] / df_clean['Impressions']
df_clean['CPC'] = df_clean['Spend'] / df_clean['Clicks']
```
