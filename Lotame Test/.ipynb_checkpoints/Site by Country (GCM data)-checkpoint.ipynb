```python
import pandas as pd
import warnings

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'C:\Users\rafael_fagundes\Downloads\1153793_L90_Days_20230619_114509_4170614186.csv', on_bad_lines='skip', skiprows=10, skipfooter=1, parse_dates=['Date'])
```

```python
df_calendar = pd.read_csv('../Helper/Calendar.csv', parse_dates=['Date'])

df_merged = pd.merge(df, df_calendar, on=['Date'], how='left')
```

```python
advertiser_groups = ['Consumer Small Business_US', 'Consumer Small Business_UK', 'Consumer Small Business_CA', 'Consumer Small Business_BR', 'Consumer Small Business_DE', 'Consumer Small Business_FR', 'Consumer Small Business_JP', 'Consumer Small Business_AU', 'Consumer Small Business_IN']
df_filtered = df_merged[df_merged['Advertiser Group'].isin(advertiser_groups)]
```

```python
df_filtered.head()
```

```python
df_grouped = df_filtered.groupby(['Advertiser','Advertiser Group','Site (CM360)'])['Impressions'].sum().reset_index()
```

```python
df_grouped.to_csv('csv/BR_CSB_Site.csv')
```
