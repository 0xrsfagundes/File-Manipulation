```python
import pandas as pd
import re

pd.set_option('max_colwidth', None)
```

```python
df_ns_audiences = pd.read_csv('csv/audiences_for_Dell_17407_.csv')
```

```python
df_ns_audiences.head()
```

```python
import pandas as pd
import re

pattern1 = r'\[(.*?)\]'

# Extract items from 'Audience_Expression' column using regex pattern
extracted_items1 = df_ns_audiences['Audience_Expression'].astype(str).str.extractall(pattern1)[0]

# Create a DataFrame to count the extracted items
df_taxonomies = pd.DataFrame(extracted_items1.value_counts()).reset_index()

# Rename the columns
df_taxonomies.rename(columns={'index':'Taxonomy', 0:'Count'}, inplace=True)

# Add columns to store the audiences containing the split taxonomies and their corresponding IDs
df_taxonomies['Audiences'] = ''
df_taxonomies['Audience_Id'] = ''

# Iterate over each row in df_taxonomies
for index, row in df_taxonomies.iterrows():
    taxonomy = row['Taxonomy']
    
    # Filter df_ns_audiences for audiences containing the taxonomy
    matching_audiences = df_ns_audiences[df_ns_audiences['Audience_Expression'].str.contains(re.escape(taxonomy))]
    
    # Concatenate the matching audience names and assign them to the 'Audiences' column in df_taxonomies
    audience_names = ', '.join(matching_audiences['Audience_Name'].astype(str))
    df_taxonomies.at[index, 'Audiences'] = audience_names
    
    # Retrieve the corresponding audience IDs and assign them to the 'Audience_Id' column in df_taxonomies
    audience_ids = ', '.join(matching_audiences['Audience_Id'].astype(str))
    df_taxonomies.at[index, 'Audience_Id'] = audience_ids

# Print the updated df_taxonomies
df_taxonomies
```

```python
df_taxonomies[['lvl1','lvl2','lvl3','lvl4','lvl5','lvl6','lvl7','lvl8']] = df_taxonomies['Taxonomy'].str.split('[>/]',expand=True, n=7)
```

```python
data_providers = '(180bytwo|33across|adadvisor|analyticsiq|clickagy|comscore|equifax|factual|lotame|startio|webbula)'

df_taxonomies['DataProvider'] = df_taxonomies['Taxonomy'].str.lower().str.extract(data_providers)
```
