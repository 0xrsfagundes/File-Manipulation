```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv(r'csv\df.csv', parse_dates=['Date'])

df = df[(df['Provider'] == 'LR') & (df['Country'] == 'BR')]
```

```python
# Split the string into a list of elements
df['split'] = df['Placement'].str.split('_')

# Determine the maximum number of split elements in any row
max_elements = max(df['split'].apply(len))

# Create new columns based on the maximum number of elements
for i in range(max_elements):
    df[f'Col{i + 1}'] = df['split'].apply(lambda x: x[i] if len(x) > i else None)

# Drop the 'split' column if needed
df = df.drop(columns=['split'])
```

```python
ax = sns.boxplot(data=df.sort_values(by='Clicks'), x='Col4',y='Clicks')

ax.set_xlabel('Placement Dimension')

# Rotate x-axis labels to 90 degrees
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

# Show the plot
plt.show()
```
