```python
import warnings
import pandas as pd
import numpy as np

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
csv_file = "csv/FY24Q3 Audience Type.csv"
df = pd.read_csv(csv_file)
```

```python
#Get Claravine ID: _d1n4f7g4t1i_
df['claravine_id'] = df['Media Buy Name'].replace('-', '_', regex=True).str.extract('(_d[a-zA-Z0-9]{10}_)')

df['claravine_id'] = df['claravine_id'].str.replace('_', '')

df.fillna(0, inplace=True)

df['Audience ID Exists'] = df['claravine_id'].apply(lambda x: 'No ID' if x == 0 else 'Have ID')
```

```python
df.head()
```

```python
df.to_csv('test.csv')
```

```python
df['Spend'] = df['Spend'].astype(str).str.replace('[\$,]', '', regex=True).astype(float)
df['Net Rev'] = df['Net Rev'].astype(str).str.replace('[\$,]', '', regex=True).astype(float)


grouping_cols = ['Audience ID Exists', 'Audience Type', 'Tactic', 'BU']

agg_funcs = {
    'Spend': 'sum',
    'Net Rev': 'sum'
}


grouped_df = df.groupby(grouping_cols).agg(agg_funcs).reset_index()


grouped_df['ROAS'] = grouped_df['Net Rev'] / grouped_df['Spend']


grouped_df = grouped_df.round(decimals=2, out=None)


grouped_df.to_csv('csv/Campaign KPIs by Audience Type export.csv')
```
