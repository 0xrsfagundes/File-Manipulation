```python
import warnings
import pandas as pd
import numpy as np

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', 100)
pd.set_option('display.width', None)
```

```python
df = pd.read_csv('csv/test.csv')
```

```python
df.head()
```

```python
# Group the data by 'Segment ID' and 'Fiscal Quarter' and count occurrences
df_filtered = df[(df['Segment ID'] != '0')][['Fiscal Quarter', 'Segment ID']]


# Save the grouped data to a new CSV file or update the existing file
df_filtered.drop_duplicates(inplace=True)
```

```python
df_filtered.head()
```

```python
# Create a new column 'Status' to initially mark all segments as 'New'
df_filtered['Status'] = 'New'

# Iterate through each unique 'Segment ID'
for segment_id in df_filtered['Segment ID'].unique():
    # Find the rows where 'Segment ID' matches the current ID
    segment_rows = df_filtered[df_filtered['Segment ID'] == segment_id]
    
    # Check if there is more than one row for this segment (i.e., it was reused)
    if len(segment_rows) > 1:
        # Set 'Status' to 'Reused' for all rows except the first one
        df_filtered.loc[segment_rows.index[1:], 'Status'] = 'Reused'
```

```python
df_filtered.head()
```

```python
df_claravine = pd.read_csv(r'C:\Users\Rafael_Fagundes\Downloads\Claravine.csv')
```

```python
df_final = pd.merge(df_filtered, df_claravine, on ='Segment ID', how='left')
```

```python
df_final.to_csv('csv/number_of_audiences.csv')
```
