```python
df_merged['audience_claravine_flag'] = df_merged['claravine_id'].apply(lambda x: 'Claravine ID' if pd.notnull(x) else 'No ID')

df_merged['audience_sfas_flag'] = df_merged['salesforce_id'].apply(lambda x: 'SFAS ID' if pd.notnull(x) else 'No ID')

df_merged['audience_flag'] = np.where(df_merged['audience_claravine_flag'] == 'Claravine ID', 'Claravine ID',
                                      np.where(df_merged['audience_sfas_flag'] == 'SFAS ID', 'SFAS ID', 'No ID'))
```

```python
df_merged['audience_ID'] = np.where(pd.notnull(df_merged['claravine_id']), df_merged['claravine_id'], df_merged['salesforce_id'])
```
