```python
import pandas as pd
import numpy as np
```

### Impressions file

```python
strFilter = 'AIQ-TEST|LIVERAMP-TEST|ZEOTAP-TEST'

df_impressions = pd.read_csv(r'C:\Users\rafael_fagundes\OneDrive - Dell Technologies\Desktop\New folder\impressions.csv', engine='python', skiprows=12,skipfooter=1, parse_dates=['Date'])

df_impressions_filtered = df_impressions[df_impressions['Placement'].str.contains(strFilter)]

df_impressions_group = df_impressions_filtered.groupby(['Date','Placement'])['Impressions','Clicks'].sum().reset_index()

df_impressions_group.info()
```

### Reach file

```python
df_reach = pd.read_csv(r'C:\Users\rafael_fagundes\OneDrive - Dell Technologies\Desktop\New folder\reach.csv', engine='python', skiprows=12, skipfooter=1, parse_dates=['Date'])
df_reach['Date']= pd.to_datetime(df_reach['Date'], errors='coerce')

df_reach.replace('-', 0, inplace=True)

df_reach_filtered = df_reach[df_reach['Placement'].str.contains(strFilter)]

df_reach_filtered.drop(columns=['Country','Advertiser','Unique Reach: Click Reach','Unique Reach: Total Reach (Co-Viewed)'], inplace=True)

df_reach_filtered.rename(columns={'Unique Reach: Impression Reach (Co-Viewed)':'Reach','Unique Reach: Average Impression Frequency (Co-Viewed)':'Frequency'}, inplace=True)

df_reach_filtered['Reach'] = df_reach_filtered['Reach'].astype(float)
df_reach_filtered['Frequency'] = df_reach_filtered['Frequency'].astype(float)

df_reach_group = df_reach_filtered.groupby(['Date','Placement'])['Reach','Frequency'].sum().reset_index()

df_reach_group.info()
```

### Visits file

```python
df_visits = pd.read_csv(r'C:\Users\rafael_fagundes\OneDrive - Dell Technologies\Desktop\New folder\visits.csv', engine='python', parse_dates=['Day'])

df_visits.rename(columns={'Day':'Date', 'Placement_Name':'Placement'}, inplace=True)

df_visits.replace('-', 0, inplace=True)

df_visits.head()

df_visits_filtered = df_visits[df_visits['Placement'].str.contains(strFilter)]

df_visits_group = df_visits_filtered.groupby(['Date','Placement'])['Visits','Unique Visitors','Total Visits'].sum().reset_index()

df_visits_group.info()
```

### Merge files

```python
df_final = pd.merge(df_impressions_group, df_reach_group, on=['Date','Placement'], how='left')

df_final = pd.merge(df_final, df_visits_group, on=['Date','Placement'], how='left')

df_final.replace(np.nan,0, inplace=True)

df_final['Vendor'] = df_final['Placement'].str.extract('(AIQ|LIVERAMP|ZEOTAP)')
    
df_final.head()
```

```python
df_calendar = pd.read_csv(r'C:\Users\rafael_fagundes\OneDrive - Dell Technologies\Desktop\New folder\calendar.csv',parse_dates=['Date'])

df_calendar.head()
```

```python
df_final = pd.merge(df_final,df_calendar,on='Date',how='left')

df_final.head()
```
